Example for class

```python
class Animal:
    def walk(self):
        return "Walking!"
    
class Dog(Animal):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        return(self.name + " Wooff!")
    
roger = Dog("roger", 8)

print(roger.name)
print(roger.age)
print(roger.bark())
print(roger.walk())
```

Nested functions example

```python
#Nested functiom

def talk(phrase):
    def say(word):
        print(word)
    
    words = phrase.split(" ")
    for word in words:
        say(word)

sentence = "I am going to london next week"

talk(sentence)
```

Using a Non Local variable - Closure

```python
def count():
    count = 0

    def increment():
        nonlocal count
        count += 1
        print(count)
    
    return increment

counter = count()
counter()
counter()
counter()
```

Ecnacpsulation

```python
class employee:
    def __init__(self, name, salary):
        self.name=name
        self.salary= salary

    @property
    def salary(self):
        return self.__salary
    
    @salary.setter
    def salary(self, value):
        if value < 0:
            raise ValueError ("Salary can't be negative..")
        self.__salary = value
    
    def increment(self, value):
        if value > 0:
            self.__salary += value

    def decrement(self, value):
        dec_value = self.__salary - value
        if dec_value <  0:
            raise ValueError ("Salary should not go below 0")
        self.__salary -= value
    
e1 = employee("aksar", 5000)
print(e1.salary)

e1.decrement(4000)
print(e1.salary)
```

Create a wallet system using encapsulation

```python
class Wallet:
    def __init__(self, owner, balance):
        self.owner  = owner
        self.balance = balance

    @property
    def balance(self):
        return self.__balance
    
    @balance.setter
    def balance(self, value):
        if value < 0:
            raise ValueError ("Balance cannot be negative")
        self.__balance = value

    def spend (self, value):
        if value < 0:
            raise ValueError ("You can't spend in negative value...")
        if value > self.balance:
            raise ValueError ("Insufficient balance")
        self.__balance -= value

    def deposit(self, value):
        if value <= 0:
            raise ValueError ("Deposit amount can't be negative..,")
        self.__balance += value
        
    def transfer(self, wallet_name, value):
        if value < 0:
            raise ValueError ("Can't transfer negative amount...")
        if value > self.__balance:
            raise ValueError ("Insufficient balance")
        self.spend(value)
        wallet_name.deposit(value)

    def __str__(self):
        return f"{self.owner} : {self.balance}"   

w1 = Wallet("Aksar", 10000)
w2 = Wallet("Ajmal", 5000)

print(w1.balance)
print(w2.balance)

w1.transfer(w2, 2000)

print(w1.balance)
print(w2.balance)
```

Class Inheritance 

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display(self):
        return f'{self.name} gets {self.salary} as salary'
    
class Developer(Employee):
    def __init__(self, name, salary, lang):
        super().__init__(name, salary, lang)
        self.lang = lang

    def code(self):
        return f'{self.name} is coding in {self.lang}'

    def display(self):
        return f'{self.name} is a developer who gets {self.salary}'

class Manager(Employee):
    def manage(self, team_size):
        self.team_size = team_size

q1 = Developer("Aksar", 30000)

q1.code("Python")

print(q1.display())
print(q1.lang)
```
