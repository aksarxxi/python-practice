Example for class

```python
class Animal:
    def walk(self):
        return "Walking!"
    
class Dog(Animal):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        return(self.name + " Wooff!")
    
roger = Dog("roger", 8)

print(roger.name)
print(roger.age)
print(roger.bark())
print(roger.walk())
```
