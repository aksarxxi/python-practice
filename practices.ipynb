Example for class

```python
class Animal:
    def walk(self):
        return "Walking!"
    
class Dog(Animal):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        return(self.name + " Wooff!")
    
roger = Dog("roger", 8)

print(roger.name)
print(roger.age)
print(roger.bark())
print(roger.walk())
```

Nested functions example

```python
#Nested functiom

def talk(phrase):
    def say(word):
        print(word)
    
    words = phrase.split(" ")
    for word in words:
        say(word)

sentence = "I am going to london next week"

talk(sentence)
```

Using a Non Local variable - Closure

```python
def count():
    count = 0

    def increment():
        nonlocal count
        count += 1
        print(count)
    
    return increment

counter = count()
counter()
counter()
counter()
```
