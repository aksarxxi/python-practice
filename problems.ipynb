To  check the number is prime

```python
def is_prime(n):
    if n <= 1:
        return "Not a prime"
    
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return "Not a prime"
    return "Prime"

print(is_prime(int(input("Enter a number to check prime : "))))
```

To check  the genere of a num

```python
def is_even(num):
    return "Even" if num % 2 == 0 else "Odd"

num = 10

print(is_even(num))
```

Largest of 3 Numbers

```python
def largest(x, y, z):
    if x > y and x > z:
        return f"{x} is the largest number"
    elif y > x and y > z:
        return f"{y} is the largest number"
    else:
        return f"{z} is the largest number"

num = input().split(",")
print(num)

num1 = int(num[0])
num2 = int(num[1])
num3 = int(num[2])

largest_num = largest(num1, num2, num3)
print(largest_num)
```

Simple Calculator

```python
def calculate(a, b, c):
    if c == "+":
        return a + b
    elif c == "-":
        return a - b
    elif c == "*":
        return a * b
    elif c == "/":
        return a / b
    return "Enter a valid input"

x, y, z = map(str, input().split(","))

print(x,z,y)

x = int(x)
y = int(y)

print(calculate(x, y, z))
```
