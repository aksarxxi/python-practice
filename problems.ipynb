Reverse a string

```python
#to reverse a string
str1 =  "Reverse this string"
rev_str = ""
for i in str1:
    rev_str = i+rev_str
print(rev_str)
```

To  check the number is prime

```python
def is_prime(n):
    if n <= 1:
        return "Not a prime"
    
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return "Not a prime"
    return "Prime"

print(is_prime(int(input("Enter a number to check prime : "))))
```

To check  the genere of a num

```python
def is_even(num):
    return "Even" if num % 2 == 0 else "Odd"

num = 10

print(is_even(num))
```
