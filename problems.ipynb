```python
def is_prime(n):
    if n <= 1:
        return "Not a prime"
    
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return "Not a prime"
    return "Prime"

print(is_prime(int(input("Enter a number to check prime : "))))
```
