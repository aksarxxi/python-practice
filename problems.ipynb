To  check the number is prime

```python
def is_prime(n):
    if n <= 1:
        return "Not a prime"
    
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return "Not a prime"
    return "Prime"

print(is_prime(int(input("Enter a number to check prime : "))))
```

Largest of 3 Numbers

```python
def largest(x, y, z):
    if x > y and x > z:
        return f"{x} is the largest number"
    elif y > x and y > z:
        return f"{y} is the largest number"
    else:
        return f"{z} is the largest number"

num = input().split(",")
print(num)

num1 = int(num[0])
num2 = int(num[1])
num3 = int(num[2])

largest_num = largest(num1, num2, num3)
print(largest_num)
```

Simple Calculator

```python
def calculate(a, b, c):
    if c == "+":
        return a + b
    elif c == "-":
        return a - b
    elif c == "*":
        return a * b
    elif c == "/":
        return a / b
    return "Enter a valid input"

x, y, z = map(str, input().split(","))

print(x,z,y)

x = int(x)
y = int(y)

print(calculate(x, y, z))
```

Palindrome Number

```python
def is_pal(number):
    number = str(number)
    if number == number[::-1]:
        return "Palindrome"
    return "Not a palindrome"

print(is_pal("Tamil"))
```

Fibonacci Series

```python
def fibo(num):
    a, b = 0, 1
    series = []
    for _ in range(num):
        series.append(a)
        a,b = b, a+b
    return series

print(fibo(15))
```

```python
def fibo_recurssive(n):
    if n <= 0:
        return 1
    return fibo_recurssive(n-1)+fibo_recurssive(n-2)

def fibonacci_series(n):
    result = []
    for i in range(n):
        result.append(fibo_recurssive(i))
    return result

print(fibonacci_series(10))
```

```python
def is_armstrong(number):
    temp = number
    length = len(str(number))
    check_arms =  0

    while temp > 0:
        check_arms += (temp%10)**length
        temp = temp//10
    return check_arms

armstrong_number = []

for i in range(100, 2000):
    if i == is_armstrong(i):
        armstrong_number.append(i)

print(armstrong_number)
```

Prime Number

```python
def is_primenum(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

prime_numbers =  []

for i in range(2, 100):
    if is_primenum(i):
        prime_numbers.append(i)

print(prime_numbers)
```
